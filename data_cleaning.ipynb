```python
import sqlite3
import pandas as pd
import re

con = sqlite3.connect("data.db")
```

## wp_posts

```python
df = pd.read_sql_query("SELECT * FROM  wp_posts", con)
df.head(5)
```

```python
df.columns
```

```python
df['post_content']
```

```python
df_new = df.copy()
for i in range(len(df)):
    result = re.findall(r"\[([^.]*?)\]([^\[\n]+)", df.loc[i, 'post_content'])
    if (len(result)):
        for key, value in result:
            df_new.loc[i, key] = value
    else:
        df_new.loc[i, 'remarks'] = df.loc[i, 'post_content']
        
```

```python
print('old wp_posts contains', len(df.columns), 'features')
print('new wp_posts contains', len(df_new.columns), 'features')
```

```python
print("dim of df is", df_new.shape)
```

```python
temp = df_new.isna().sum()
temp[temp != 0]
```

```python
df_new.describe()
```

```python
df_new.drop(['post_content'], axis=1)
df_new.to_csv('ws_post.csv' ,index=False)
```

## wp_tagthis

```python
df_t = pd.read_sql_query("SELECT * FROM  wp_tagthis", con)
df_t.head(5)
```

```python
df_tnew = df_t.copy()
for i in range(len(df_t)):
    result= re.findall(r"\[([^.]*?)\]([^\[\n]+)", df_t.loc[i, 'tag'])
    if (len(result)):
        id, name = result[0]
        df_tnew.loc[i, 'name'] = name
        for pair in id.split('%%'):
            key, value = pair.split('=')
            df_tnew.loc[i, key] = value
df_tnew.head(5)
```

```python
print("dim of ws_tagthis is", df_tnew.shape)
temp = df_tnew.isna().sum()
temp[temp != 0]
```

```python
df_tnew.describe()
```

```python
df_tnew.to_csv('ws_tagthis.csv', index=False)
```

## New table

### album

- albumID => int increment
- albumName => 'name' from wp_ngg_gallery
- aTitle => 'post_title' of wp_posts
- thumbnail => first pic of album
- aDate => 'post_date_gmt' of wp_posts
- mDate => 'post_modified_gmt' of wp_posts
- aURL => 'guid' from wp_posts

```python
df_gallery = pd.read_sql_query("SELECT * FROM  wp_ngg_gallery", con)
df_gallery.head(5)
```

```python
def checkna(a, b):
    if a.values:
        return a.values
    return b

df = df_gallery['name'].drop_duplicates()
df = df.reset_index().rename(columns={'index':'albumID'})

df_post = pd.read_csv('wp_post.csv')

for i in range(len(df)):
    temp = df_post[df_post['post_name'] == df.loc[i, 'name']]
    if not temp.empty:
        df.loc[i, 'aTitle'] = temp['evtitle'].values
        df.loc[i, 'aDate'] = temp['post_date_gmt'].values
        df.loc[i, 'mDate'] = temp['post_modified_gmt'].values
        df.loc[i, 'aURL'] = temp['guid'].values
df
```

```python
df.isna().sum()
```

```python
df.to_csv('./newdb/album.csv', index=False)
```

### photo

- photoID => int increment
- filename => 'image_slug' of wp_ngg_pictures
- imagepath => 'name' + 'image_path'
- physicalFormat ???
- archiveBox ???
- aID => 'albumID' of album

```python
df_gal = pd.read_sql_query("SELECT * FROM  wp_ngg_gallery", con)

df_pic = pd.read_sql_query("SELECT * FROM  wp_ngg_pictures", con)
df_pic.head(1)
```

```python
df = df_pic['image_slug'].drop_duplicates()
df = df.reset_index().rename(columns={'index':'photoID', 'image_slug':'filename'})

df_gallery = pd.read_csv('newdb/album.csv', index_col=0)

for i in range(len(df)):
    temp = df_gal[df_gal['gid'] == df_pic[df_pic['image_slug'] == df.loc[i, 'filename']]['galleryid'].values[0]]
    if not temp.empty:
        df.loc[i, 'imagepath'] = temp['path'].values[0] + '/' + df_pic[df_pic['image_slug'] == df.loc[i, 'filename']]['filename'].values
        df.loc[i, 'aID'] = df_gallery[df_gallery['name'] == temp['name'].values[0]]['albumID'].values[0]
df.head(1)
```

```python
df.to_csv('newdb/photo.csv', index=False)
```

### tag

- ptagID => int increment
- phototID => 'photoID' of photo
- personID => 'personID' of personID
- c1 => 'posLeft' of wp_tagthis
- c2 => 'posTop' of wp_tagthis
- c3 => 'width' of wp_tagthis
- c4 => 'height' of wp_tagthis

```python
df_tag = pd.read_csv('wp_tagthis.csv')

df_person = pd.read_csv('newdb/tagPerson.csv')

df_photo = pd.read_csv('newdb/photo.csv')

df_p = pd.read_sql_query("SELECT * FROM  wp_ngg_pictures", con)

df_album = pd.read_csv('newdb/album.csv')

df_tag.head(1)
```

```python
df = df_tag.loc[:,['p','pid','name','posLeft','posTop','width','height']]
df = df.reset_index().rename(columns={'p':'albumID', 'pid':'photoID', 'name':'personID', 'posLeft':'c1', 'posTop':'c2', 'width':'c3', 'height':'c4'}).drop(columns=['index'], axis=1)
df.index.names = ['tagID']
for i in range(len(df)):
    if len(df_person[df_person['name'] == df.loc[i, 'personID']]['personID'].values):
        df.loc[i , 'personID'] = df_person[df_person['name'] == df.loc[i, 'personID']]['personID'].values
    if len(df_album[df_album['name'] == df.loc[i, 'albumID']]['albumID'].values):
        df.loc[i, 'albumID'] = df_album[df_album['name'] == df.loc[i, 'albumID']]['albumID'].values 
df
```

```python
df.to_csv('newdb/tag.csv')
```
